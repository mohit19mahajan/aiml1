```python
## 1. Read the data set and replace dashes with 0 to make sure you can perform arithmetic
##operations on the data. (5 points)

ds.replace("-",0,inplace=True)
```

```python
## 2. Print all the teams which have started playing between 1930-1980. Use “Debut” column
## (5 points)

ds[(ds["Debut"].astype(str).str[0:4].astype(int) > 1929) & (ds["Debut"].astype(str).str[0:4].astype(int) <= 1980)]
```

```python
ds["PointsInt"]=ds.Points.astype(int)
```

```python
##3. Print the list of teams which came Top 5 in terms of points (2.5 points)
ds.sort_values("PointsInt",ascending=False).iloc[0:5]
```

```python
## 4. Write a function with name “Goal_diff_count” which should return all the teams with their
## Goal Differences. Using the same function, find the team which has maximum and
## minimum goal difference. (5 points)

def Goal_diff_count():
    ds['goalsdiff'] = ds['GoalsFor'].astype(int) - ds['GoalsAgainst'].astype(int)
    return ds[['Team','goalsdiff']]

Goal_diff_count()
```

```python
max_goals = ds[Goal_diff_count()['goalsdiff'] == np.max(Goal_diff_count()['goalsdiff'])]
max_goals
```

```python
min_goals = ds[Goal_diff_count()['goalsdiff'] == np.min(Goal_diff_count()['goalsdiff'])]
min_goals
```

```python
## 5. Create a new column with the name “Winning Percent” and append it to the data set (5
## points)
ds["WinningPercent"] = ds["GamesWon"].astype(int)/ds["GamesPlayed"].astype(int) * 100
ds["WinningPercent"].fillna(0)
ds.sort_values("WinningPercent",ascending=False).iloc[0:5]
```

```python
## 6. Group teams based on their “Best position” and print the sum of their points for all
## positions (7.5 points)

ds.groupby(by = "BestPosition")["PointsInt"].agg("sum")
```
