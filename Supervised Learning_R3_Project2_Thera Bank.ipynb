```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.neighbors import KNeighborsClassifier

from sklearn.linear_model import LogisticRegression

from sklearn.naive_bayes import GaussianNB
```

```python
X = cust.drop("Personal Loan",axis=1)
y = cust["Personal Loan"]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.30,random_state = 10 )
```

```python
sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
# applying knn algorithm

knnmodel = KNeighborsClassifier(n_neighbors=5,metric="euclidean")
knnmodel.fit(X_train_scaled,y_train)
predictedknn = knnmodel.predict(X_test)
scoreknn = knnmodel.score(X_test,y_test)
print(scoreknn)
print(confusion_matrix(y_test,predictedknn))
#print(classification_report(y_test,predictedknn))
```

```python
# Logistic regression
logisticmodel = LogisticRegression()
logisticmodel.fit(X_train_scaled,y_train)
predictedlogi = logisticmodel.predict(X_test)
scorelogi = logisticmodel.score(X_test,y_test)
print(scorelogi)
print(confusion_matrix(y_test,predictedlogi))
#print(classification_report(y_test,predictedknn))
```

```python
# Naive bayes regression
NBmodel = GaussianNB()
NBmodel.fit(X_train_scaled,y_train)
predictedNB = NBmodel.predict(X_test)
scoreNB = NBmodel.score(X_test,y_test)
print(scoreNB)
print(confusion_matrix(y_test,predictedNB))
#print(classification_report(y_test,predictedknn))
```
